```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Importing the dataset
url = 'https://raw.githubusercontent.com/andrewbeattycourseware/PFDA-courseware/main/code/data/projectedbirths-cso.csv'
data = pd.read_csv(url)

print(data.columns)
```

```python
plt.plot(data['Year'], data['VALUE'], marker='o')
plt.title('Forecast for births in Ireland')
plt.xlabel('Year')
plt.ylabel('VALUE')
plt.show()
```
