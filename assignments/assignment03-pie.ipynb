```python
# Import necessary libraries

import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# Load the dataset into a DataFrame 
url = "https://drive.google.com/uc?id=1AWPf-pJodJKeHsARQK_RHiNsE8fjPCVK&export=download"
```

```python
# Read the CSV file from the URL
df = pd.read_csv(url)
```

```python
# Extract domain from email and split
df['domain'] = df['Email'].str.split('@').str[1] 

# Count occurrences of each domain
domain_counts = df['domain'].value_counts()
print(domain_counts)
```

```python
# Make each slice separately.
# https://matplotlib.org/stable/gallery/pie_and_polar_charts/pie_features.html#explode-shade-and-rotate-slices
explode = [0.1, 0.1, 0.1]  

domain_counts.plot.pie(
    figsize=(5, 5),
    title='Email Domain Distribution',
    explode=explode,
    autopct='%1.1f%%',
    startangle=130)

plt.xlabel('')
plt.ylabel('') 
plt.show()  
```
