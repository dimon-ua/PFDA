```python
import re
```

```python
regex = r"[0-9]+"
```

```python
# getting access to /log-file/access.log.txt 
with open("log-file/access.log.txt") as f:
    log_data = f.read()

print(log_data)
```

```python
# serching for ip addresses
matches = re.findall(r"\d+\.\d+\.\d+\.\d+", log_data)
print(matches)
```

```python
# searching for dates
date_matches = re.findall(r"\d{2}/[A-Za-z]{3}/\d{4}", log_data)
print(date_matches) 
```
